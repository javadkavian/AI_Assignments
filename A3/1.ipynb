```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

# Input 2D points
points = np.array([[2, 10], [2, 5], [8, 4], [5, 8], [7, 5], [6, 4], [1, 2], [4, 9]])

# Run DBSCAN
dbscan = DBSCAN(eps=np.sqrt(10), min_samples=2)
cluster_labels = dbscan.fit_predict(points)

# Plot the scatter plot
plt.figure(figsize=(8, 6))
for label in np.unique(cluster_labels):
    if label == -1:
        # Noise points (not assigned to any cluster)
        plt.scatter(points[cluster_labels == label][:, 0], points[cluster_labels == label][:, 1], color='gray', label="Noise")
    else:
        plt.scatter(points[cluster_labels == label][:, 0], points[cluster_labels == label][:, 1], label=f"Cluster {label}")

plt.xlabel("X")
plt.ylabel("Y")
plt.title("DBSCAN Clustering")
plt.legend()
plt.grid(True)
plt.show()
```
